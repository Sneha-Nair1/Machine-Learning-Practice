```python
data1['tweet']=data1['tweet'].apply(lambda x: x.split())
data1.head()
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
data1['tweet']=data1['tweet'].apply(lambda x:[stemmer.stem(i) for i in x])
data1.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stopwords=nltk.corpus.stopwords.words('english')
newStopwords=['u','go','got','via','or','ur','us','in','i','let','the','to','is','amp','make','one','day','days','get']
stopwords.extend(newStopwords)
```

```python
import string
def process(text):
  nopunc=set(char for char in list(text)if char not in string.punctuation)
  nopunc=" ".join(nopunc)
  return[word for word in nopunc.lower().split() if word.lower()not in stopwords]
```

```python
data1['tweet']=data1['tweet'].apply(process)
data1.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
words=[]
for line in data1['tweet']:
  words.extend(line)
wordfreq=Counter(words)
wordcloud=WordCloud(
    background_color='#000000',
    max_words=2000,
    stopwords=stopwords
    ).generate_from_frequencies(wordfreq)
plt.figure(figsize=(10,9))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
def string(text):
  to_return=""
  for i in list(text):
    to_return += str(i) + " "
  to_return = to_return[:-1]
  return to_return
data1['tweet']= data1['tweet'].apply(string)
data1.head()
```
